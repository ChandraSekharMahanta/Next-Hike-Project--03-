# Visualize Checking Outliers Using Box Plot

```python
num_cols = df.select_dtypes(include=["float64", "int64"]).columns
plt.figure(figsize=(24, 19))
for i, col in enumerate(num_cols, 1):
    plt.subplot(7, 6, i)  
    sns.boxplot(y=df[col])
plt.title(col)
plt.tight_layout()
plt.show()
```

```python
for i in num_cols:
    Q1 = df[i].quantile(0.25)
    Q2 = df[i].quantile(0.50)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    
    LL = Q1 - 1.5 * IQR  
    UL = Q3 + 1.5 * IQR  
    
    n = df[(df[i] < LL) | (df[i] > UL)].count()[0]  
    print(f"The first quartile for the column {i} is {Q1}")
    print(f"The second quartile (median) for the column {i} is {Q2}")
    print(f"The third quartile for the column {i} is {Q3}")
    print(f"The interquartile range (IQR) is {IQR}")
    print(f"The lower limit for the column {i} is {LL}")
    print(f"The upper limit for the column {i} is {UL}")
    print(f"The number of outliers for the column {i} is {n}")
    
```

# Replace Outliers with Median

```python
df[i] = np.where((df[i] < LL) | (df[i] > UL), df[i].median(), df[i])
n = df[(df[i] < LL) | (df[i] > UL)].count()[0]
    
print("The number of outliers after imputation is", n)
print("Minimum value in column after imputation:", df[i].min())
print("Maximum value in column after imputation:", df[i].max())
print("Median value in column after imputation:", df[i].median())
```

# Market Trends and Historical Pricing

```python
price_trends = df.groupby("YrSold")["SalePrice"].mean()
plt.figure(figsize=(8, 5))
sns.lineplot(x=price_trends.index, y=price_trends.values, marker="o")
plt.xlabel("Year Sold")
plt.ylabel("Average Sale Price")
plt.title("Market Trends: Average House Price Over Time")
plt.show()
```

```python
# The average sale price increased from 2006 to 2007, reaching its peak in 2007 (~$186,000).
# After 2007, there was a sharp decline in 2008, likely due to the global financial crisis (2008 housing market crash).
# A slight recovery is visible in 2009, where prices increased again.
# However, in 2010, the decline continued, showing that the market had not fully recovered.
```

```python
price_trends = df.groupby("MoSold")["SalePrice"].mean()
plt.figure(figsize=(8, 5))
sns.lineplot(x=price_trends.index, y=price_trends.values, marker="o")
plt.xlabel("MoSold")
plt.ylabel("Average Sale Price")
plt.title("Market Trends: Average House Price Over Time")
plt.show()
```

```python
# The house prices fluctuate throughout the year, indicating seasonal variations in the housing market.
# Lowest prices are observed in April and May, suggesting a potential dip in demand during these months.
# Highest prices occur in September and November, which might indicate peak demand seasons.
# There are multiple ups and downs, showing that house prices do not follow a consistent pattern but are affected by seasonal trends,
   market conditions, and buyer behavior.
```
